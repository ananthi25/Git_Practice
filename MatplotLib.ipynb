```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.scatter(x,y)  #scatter
plt.plot(x,y)
plt.show()        #For Showing Graph
```

```python
x=[1,2,3,4,5]
x2=[1,2,3,4,5]
y=[1,2,3,4,5]
y2=[5,4,3,2,1]

plt.scatter(x,y)
plt.scatter(x2,y2)
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

Adding Colors
* 'r'  -red
* 'g'  -green
* 'b'  -blue
* 'y'  -yellow
* 'c'  -cyan
* 'k'  -black
* 'w'  -white

```python
x=[1,2,3,4,5]
x2=[1,2,3,4,5]
y=[1,2,3,4,5]
y2=[5,4,3,2,1]

plt.plot(x,y,"r")       #red color
plt.plot(x2,y2)
plt.show()
```

Line Style 
* '-'  -solid line
* '--' -dashed line
* '-.' -dash-dot line
* ':'  -dotted line

```python
x=[1,2,3,4,5]
x2=[1,2,3,4,5]
y=[1,2,3,4,5]
y2=[5,4,3,2,1]

plt.plot(x,y,"r--")       #red color double dash
plt.plot(x2,y2)
plt.show()
```

```python
x=[1,2,3,4,5]
x2=[1,2,3,4,5]
y=[1,2,3,4,5]
y2=[5,4,3,2,1]

plt.plot(x,y,"r-.")       #red color dash dot
plt.plot(x2,y2)
plt.show()
```

```python
x=[1,2,3,4,5]
x2=[1,2,3,4,5]
y=[1,2,3,4,5]
y2=[5,4,3,2,1]

plt.plot(x,y,"r:")       #red color
plt.plot(x2,y2)
plt.show()
```

Marker Style
* 's'  -square marker
* 'p'  -pentagon marker
* '*'  -star marker
* 'h'  -hexagon1 marker
* 'H'  -hexagon2 marker
* '+'  -Plus marker
* 'x'  -X marker
* 'D'  -diamond marker

```python
x=[1,2,3,4,5]
x2=[1,2,3,4,5]
y=[1,2,3,4,5]
y2=[5,4,3,2,1]

plt.plot(x,y,"rx")       #red color
plt.plot(x2,y2)
plt.show()
```

```python
x=[1,2,3,4,5]
x2=[1,2,3,4,5]
y=[1,2,3,4,5]
y2=[5,4,3,2,1]

plt.plot(x,y,"rs--")       #red color
plt.plot(x2,y2)
plt.show()
```
