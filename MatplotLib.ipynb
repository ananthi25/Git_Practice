```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.scatter(x,y)  #scatter
plt.plot(x,y)
plt.show()        #For Showing Graph
```

```python
x=[1,2,3,4,5]
x2=[1,2,3,4,5]
y=[1,2,3,4,5]
y2=[5,4,3,2,1]

plt.scatter(x,y)
plt.scatter(x2,y2)
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

Adding Colors
* 'r'  -red
* 'g'  -green
* 'b'  -blue
* 'y'  -yellow
* 'c'  -cyan
* 'k'  -black
* 'w'  -white

```python
x=[1,2,3,4,5]
x2=[1,2,3,4,5]
y=[1,2,3,4,5]
y2=[5,4,3,2,1]

plt.plot(x,y,"r")       #red color
plt.plot(x2,y2)
plt.show()
```

Line Style 
* '-'  -solid line
* '--' -dashed line
* '-.' -dash-dot line
* ':'  -dotted line

```python
x=[1,2,3,4,5]
x2=[1,2,3,4,5]
y=[1,2,3,4,5]
y2=[5,4,3,2,1]

plt.plot(x,y,"r--")       #red color double dash
plt.plot(x2,y2)
plt.show()
```

```python
x=[1,2,3,4,5]
x2=[1,2,3,4,5]
y=[1,2,3,4,5]
y2=[5,4,3,2,1]

plt.plot(x,y,"r-.")       #red color dash dot
plt.plot(x2,y2)
plt.show()
```

```python
x=[1,2,3,4,5]
x2=[1,2,3,4,5]
y=[1,2,3,4,5]
y2=[5,4,3,2,1]

plt.plot(x,y,"r:")       #red color
plt.plot(x2,y2)
plt.show()
```

Marker Style

* '*'  -star marker
* 'h'  -hexagon1 marker
* 'H'  -hexagon2 marker

```python
import numpy as np

x=np.array([1,2,3,4,5,6])
y=2*(x**4)+100
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,10,0.1)        #arange function
y=2*(x**4)+100
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,10,0.25)        
y=x**4

plt.plot(x,y,color="red",marker="+",linewidth=2)
plt.show()
```

```python
x=np.arange(0,10,0.25)        
y=x**4
y2=x**3

plt.plot(x,y,color="red",label="x^4")
plt.plot(x,y2,color="blue",label="x^3")
plt.ylabel("y",fontsize=15)
plt.xlabel("X",fontsize=15)
plt.legend()                                   #legend function for label
plt.show()
```

```python
x=np.arange(0,10,0.25)        
y=x**4
y2=x**3

plt.plot(x,y,color="red",label="x^4")
plt.plot(x,y2,color="blue",label="x^3")
plt.ylabel("y",fontsize=15)
plt.xlabel("X",fontsize=15)
plt.legend()                                 
plt.title("Sample Title")
plt.axis([0,20,0,10000])
plt.text(10,12500,"Ananthi",fontsize=15)
plt.grid()
plt.show()
```

Pie Chart

```python
profit=[10,20,30,25]
labels=["Amazon","Flipcart","Snapdeal","Jiomart"]

colors=["red","yellow","pink","blue"]
explode=[0.05,0.1,0,0]
plt.pie(profit,colors=colors, labels=labels,autopct="%.2f%%",explode=explode,counterclock=False,startangle=90)          #pie
plt.show()
```
